```python
# 把VD編號變成 0-50的數字


dfnhcg=dfnhc.set_index(['VD編號','方向']).index.unique().tolist();dfnhcg
# 把VD編號跟方向變成Index，取出來變成list
dfvdd=pd.DataFrame(dfnhcg,columns=['a','b']);dfvdd
# 在把剛剛的list變成df column name =[a,b]
dfvdd['c']=dfvdd.index;dfvdd
# 新增一欄位C是預設index 0-50
dfvdd1=dfvdd.set_index(['a','b']);dfvdd1
#  把ab設成index
dfnhcri=dfnhc.set_index(['VD編號','方向']);dfnhcri
# 再把原始資料的VD編號及方向設成index
dfnhcri['sortby'] = None
# 新增一欄位sortby
x1 = dfnhcri['sortby']
x2 = dfvdd1['c']
# 把VD編號跟方向新增sortby 編號
x1.update(x2);x1
# 把編號更新到原來的DF上
dfnhcri['sortby']=x1.values
dfnhcris=dfnhcri.sort_values(['sortby','日期']);dfnhcris
# 重新 index 就成功拉
dfnhcris=dfnhcris.reset_index();
```

```python
#0225version
def cleanout(inp,i):
    dfvdi= inp.where(inp['sortby']==i).dropna();
    #只取想要的欄位，重新命名
    outp = dfvdi.ix[:,[12,1,4,5,6,8,9,7]].copy()
    outp.columns = ['vdnum','direct','lane','date','time','speed','accu','eqflow'];outp
    #新增欄位較dt是日期加時間
    outp['dt'] = outp.date+" "+outp.time
    #把dt變成index
    outp.index = pd.to_datetime(outp['dt']);outp
    #時間為index就可以用resample魔法！
    outp=outp.resample('H').last().fillna(np.nan);outp
    #把不太重要的空格用後面的直往前補
    outp.vdnum =outp.vdnum.bfill();outp
    outp.direct =outp.direct.bfill();outp
    outp.lane =outp.lane.bfill();outp

    #再次將dt變成index
    outp['dt'] = outp.index;outp
    #只要年的十位數跟個位數
    outp['YY'] = outp['dt'].dt.strftime('%Y').astype('int32')-2000
    #把dt變成字串形式
    outp['dt'] = outp['dt'].dt.strftime('%Y-%m-%d %H:%M:%S')
    #因為date不是值就是nan，不是值(nan)的在fakeornot得欄位為1，表示全部的值都是假的
    outp['fakeornot'] = outp.date.where(outp.date,1);outp
    #fakeornot不是1就是0，0表示都是實際值
    outp['fakeornot'] = outp['fakeornot'].where(outp['fakeornot']==1,0);outp

    #把date跟time值捕進去
    outp.date =outp.dt.str.split(' ',expand=True).ix[:,0];outp
    outp.time =outp.dt.str.split(' ',expand=True).ix[:,1];outp

    #重設index
    outp=outp.reset_index(drop=True)
    #丟掉dt
    outp=outp.drop('dt',axis =1)
    #方向變成數字
    outp.replace(to_replace='東',value=1,inplace=True );outp
    outp.replace(to_replace='西',value=2,inplace=True );outp
    outp.replace(to_replace='南',value=3,inplace=True );outp
    outp.replace(to_replace='北',value=4,inplace=True );outp
    #把date變成日期格式
    outp.date = pd.to_datetime(outp.date);
    #新增一欄看是一週中的第幾天
    outp['weekday'] = outp['date'].dt.dayofweek;outp
    #在把date 扮成一年中的第幾天
    outp.date= outp.date.dt.dayofyear;outp
    #把time變成數字
    outp.replace(to_replace='00:00:00',value=0,inplace=True );outp
    outp.replace(to_replace='01:00:00',value=1,inplace=True );outp
    outp.replace(to_replace='02:00:00',value=2,inplace=True );outp
    outp.replace(to_replace='03:00:00',value=3,inplace=True );outp
    outp.replace(to_replace='04:00:00',value=4,inplace=True );outp
    outp.replace(to_replace='05:00:00',value=5,inplace=True );outp
    outp.replace(to_replace='06:00:00',value=6,inplace=True );outp
    outp.replace(to_replace='07:00:00',value=7,inplace=True );outp
    outp.replace(to_replace='08:00:00',value=8,inplace=True );outp
    outp.replace(to_replace='09:00:00',value=9,inplace=True );outp
    outp.replace(to_replace='10:00:00',value=10,inplace=True );outp
    outp.replace(to_replace='11:00:00',value=11,inplace=True );outp
    outp.replace(to_replace='12:00:00',value=12,inplace=True );outp
    outp.replace(to_replace='13:00:00',value=13,inplace=True );outp
    outp.replace(to_replace='14:00:00',value=14,inplace=True );outp
    outp.replace(to_replace='15:00:00',value=15,inplace=True );outp
    outp.replace(to_replace='16:00:00',value=16,inplace=True );outp
    outp.replace(to_replace='17:00:00',value=17,inplace=True );outp
    outp.replace(to_replace='18:00:00',value=18,inplace=True );outp
    outp.replace(to_replace='19:00:00',value=19,inplace=True );outp
    outp.replace(to_replace='20:00:00',value=20,inplace=True );outp
    outp.replace(to_replace='21:00:00',value=21,inplace=True );outp
    outp.replace(to_replace='22:00:00',value=22,inplace=True );outp
    outp.replace(to_replace='23:00:00',value=23,inplace=True );outp
    ###漫長的補缺值的平均值  version2.0223

    eqlow,eqhigh=outp.eqflow.quantile([0.005,0.995])
    aclow,achigh=outp.accu.quantile([0.005,0.995])
    splow,sphigh=outp.speed.quantile([0.005,0.995])

    outp.eqflow=outp.eqflow.to_frame().apply(lambda x: x[(x>eqlow) & (x < eqhigh)])
    outp.accu=outp.accu.to_frame().apply(lambda x: x[(x>aclow) & (x < achigh)])
    outp.speed=outp.speed.to_frame().apply(lambda x: x[(x>splow) & (x < sphigh)])

    #當量是離群職  fakeornot = 2
    #佔有率是離群職  fakeornot = 3,
    #速度是離群職  fakeornot = 4
    #當量跟佔有率都是離群值 fakeornot = 5
    #當量跟速度都是離群值 fakeornot = 6,
    #速度跟佔有率都是離群值 fakeornot = 7
    #當量跟佔有率跟速度都是離群值 fakeornot = 8


    outp.ix[(outp.eqflow.isnull())&(outp.fakeornot==0),'fakeornot']= 2
    outp.ix[(outp.accu.isnull())&(outp.fakeornot==0),'fakeornot']= 3
    outp.ix[(outp.speed.isnull())&(outp.fakeornot==0),'fakeornot']= 4
    outp.ix[(outp.accu.isnull())&(outp.fakeornot==2),'fakeornot']= 5
    outp.ix[(outp.speed.isnull())&(outp.fakeornot==2),'fakeornot']= 6
    outp.ix[(outp.speed.isnull())&(outp.fakeornot==3),'fakeornot']= 7
    outp.ix[(outp.speed.isnull())&(outp.fakeornot==5),'fakeornot']= 8


    # 只要把time跟weekday變成index而且只要是有值是nan的就抓出來
    nandata=outp.ix[:,[4,10]].where((outp.fakeornot != 0)).dropna(how = 'all').copy().set_index(['time','weekday'])
    nandata=nandata.index.unique().tolist();nandata.sort();nandata

    #跑個回圈把應該要的平均值丟進去
    for ti,we in nandata:
    #     print ti,we
        spme=outp.where((outp.time == ti)&(outp.weekday==we)).speed.mean()
        eqme=outp.where((outp.time == ti)&(outp.weekday==we)).eqflow.mean()
        acme=outp.where((outp.time == ti)&(outp.weekday==we)).accu.mean()
    #     print spme
        #缺值要補
        outp.ix[(outp.time == ti)&(outp.weekday==we)&((outp.fakeornot==1)), 'speed']=spme
        outp.ix[(outp.time == ti)&(outp.weekday==we)&((outp.fakeornot==1)), 'eqflow']=eqme
        outp.ix[(outp.time == ti)&(outp.weekday==we)&((outp.fakeornot==1)), 'accu']=acme
        #離群值也要補 
        outp.ix[(outp.time == ti)&(outp.weekday==we)&((outp.fakeornot==2)), 'eqflow']=eqme
        outp.ix[(outp.time == ti)&(outp.weekday==we)&((outp.fakeornot==3)), 'accu']=acme
        outp.ix[(outp.time == ti)&(outp.weekday==we)&((outp.fakeornot==4)), 'speed']=spme

        outp.ix[(outp.time == ti)&(outp.weekday==we)&((outp.fakeornot==5)), 'eqflow']=eqme
        outp.ix[(outp.time == ti)&(outp.weekday==we)&((outp.fakeornot==5)), 'accu']=acme

        outp.ix[(outp.time == ti)&(outp.weekday==we)&((outp.fakeornot==6)), 'eqflow']=eqme
        outp.ix[(outp.time == ti)&(outp.weekday==we)&((outp.fakeornot==6)), 'speed']=spme

        outp.ix[(outp.time == ti)&(outp.weekday==we)&((outp.fakeornot==7)), 'speed']=spme
        outp.ix[(outp.time == ti)&(outp.weekday==we)&((outp.fakeornot==7)), 'accu']=acme

        outp.ix[(outp.time == ti)&(outp.weekday==we)&((outp.fakeornot==8)), 'accu']=acme
        outp.ix[(outp.time == ti)&(outp.weekday==we)&((outp.fakeornot==8)), 'speed']=spme
        outp.ix[(outp.time == ti)&(outp.weekday==we)&((outp.fakeornot==8)), 'eqflow']=eqme

    """#檢查是否有nan
    # outp.isnull().values.any()"""

    cols =outp.columns.tolist()
    #位置移動一下
    outp= outp[cols[:3]+cols[-3:]+cols[3:-3]];outp


    # 某個時間要把fakeornot去掉 

    outpp = outp.drop('fakeornot',axis =1).values
    outpp = outpp.astype('float32')


    train_size = int(len(outpp) * 0.75)
    test_size = len(outpp) - train_size
    train, test = outpp[0:train_size,:], outpp[train_size:len(outpp),:]
    if (len(train)%168 != 0) or (len(test)%168 != 0) :
        train=train[:-(len(train)%168)]
        test=test[:-(len(test)%168)]

    return train,test,outp.values
```

```python
Temp = []
vdlist=oli['sortby'].unique().tolist()

for ele in vdlist:
    correqe = oli.where(oli['sortby']==ele).dropna()['eqflow'].tolist()
    corrspe = oli.where(oli['sortby']==ele).dropna()['speed'].tolist()
    corrace = oli.where(oli['sortby']==ele).dropna()['accu'].tolist()
    direc = oli.where(oli['sortby']==ele).dropna()['direct'].unique()[0]
    
    eqsp =st.spearmanr(correqe,corrspe)[0]
    eqac =st.spearmanr(correqe,corrace)[0]
    spac =st.spearmanr(corrspe,corrace)[0]
    
#     print eqsp,eqac,spac,vdno,direc
    Temp.append([eqsp,spac])
    
X = np.array(Temp);

silhouette_avgs = []
ks = range(2,51)

for k in ks:
    kmeans_fit = cluster.KMeans(n_clusters = k).fit(X)
    cluster_labels = kmeans_fit.labels_
    silhouette_avg = metrics.silhouette_score(X, cluster_labels)
    silhouette_avgs.append(silhouette_avg)

# 作圖並印出 k = 2 到 50 的績效
plt.bar(ks, silhouette_avgs)
plt.show()
print(silhouette_avgs)
```

```python
def clean(inp):
    outp = inp.ix[:,[12,1,4,5,6,8,9,10,7]].copy()
    outp.columns = ['a','b','c','d','e','f','g','h','i'];outp
    outp.i=outp.i/10;outp
    outp.h=outp.h/10;outp
    outp.replace(to_replace='東',value=1,inplace=True );outp
    outp.replace(to_replace='西',value=2,inplace=True );outp
    outp.replace(to_replace='南',value=3,inplace=True );outp
    outp.replace(to_replace='北',value=4,inplace=True );outp
    outp.d = pd.to_datetime(outp.d);
    outp['weekday'] = outp['d'].dt.dayofweek;outp
    outp.d= outp.d.dt.dayofyear;outp
    outp.replace(to_replace='00:00:00',value=0,inplace=True );outp
    outp.replace(to_replace='01:00:00',value=1,inplace=True );outp
    outp.replace(to_replace='02:00:00',value=2,inplace=True );outp
    outp.replace(to_replace='03:00:00',value=3,inplace=True );outp
    outp.replace(to_replace='04:00:00',value=4,inplace=True );outp
    outp.replace(to_replace='05:00:00',value=5,inplace=True );outp
    outp.replace(to_replace='06:00:00',value=6,inplace=True );outp
    outp.replace(to_replace='07:00:00',value=7,inplace=True );outp
    outp.replace(to_replace='08:00:00',value=8,inplace=True );outp
    outp.replace(to_replace='09:00:00',value=9,inplace=True );outp
    outp.replace(to_replace='10:00:00',value=10,inplace=True );outp
    outp.replace(to_replace='11:00:00',value=11,inplace=True );outp
    outp.replace(to_replace='12:00:00',value=12,inplace=True );outp
    outp.replace(to_replace='13:00:00',value=13,inplace=True );outp
    outp.replace(to_replace='14:00:00',value=14,inplace=True );outp
    outp.replace(to_replace='15:00:00',value=15,inplace=True );outp
    outp.replace(to_replace='16:00:00',value=16,inplace=True );outp
    outp.replace(to_replace='17:00:00',value=17,inplace=True );outp
    outp.replace(to_replace='18:00:00',value=18,inplace=True );outp
    outp.replace(to_replace='19:00:00',value=19,inplace=True );outp
    outp.replace(to_replace='20:00:00',value=20,inplace=True );outp
    outp.replace(to_replace='21:00:00',value=21,inplace=True );outp
    outp.replace(to_replace='22:00:00',value=22,inplace=True );outp
    outp.replace(to_replace='23:00:00',value=23,inplace=True );outp
    cols =outp.columns.tolist()
    outp= outp[cols[:3]+cols[-1:]+cols[3:-1]];outp
    outp.columns = ['a','b','c','d','e','f','g','h','i','j'];outp
    outpp = outp.values
    outpp = outpp.astype('float32')
    
    
    train_size = int(len(outpp[:-(len(outpp)%168)]) * 0.75)
    test_size = len(outpp[:-(len(outpp)%168)]) - train_size
    train, test = outpp[:-(len(outpp)%168)][0:train_size,:], outpp[:-(len(outpp)%168)][train_size:len(outpp),:]
    
    return train, test
    
    
    
```

```python


tralist = []
teslist = []
for geg in dictvd.keys():
    temptra,temptes=clean(dictvd[geg].reset_index())
    #train
    for ele in temptra.tolist():
        tralist.append(ele)
    #test
    for ele3 in temptes.tolist():
        teslist.append(ele3)

cluster01tra=np.array(tralist)
print len(cluster01tra)
cluster01tes=np.array(teslist)
print len(cluster01tes)

    
    
```

```python
# convert an array of values into a dataset matrix
def create_dataset(dataset, look_back):
    dataX, dataY = [], []
    for i in range(0,len(dataset)-look_back-1,look_back):
        a = dataset[i:(i+look_back), 0:nort]
        dataX.append(a)
        dataY.append(dataset[i + look_back:i+look_back+look_back, nort-1])
#     return dataX,dataY
    return np.array(dataX), np.array(dataY)

# 沒有車間劇=9有就是10
nort = 10
# magic number
look_back = 168
# cut off remainder
train = cluster01tra[:-(len(cluster01tra)%168)]
test = cluster01tes[:-(len(cluster01tes)%168)]
# check data lenth when cut off remainder
print(len(train), len(test))

# seperate data to input"X" and output"Y"
trainX, trainY = create_dataset(train, look_back)
testX, testY = create_dataset(test, look_back)
    
# create and fit Multilayer Perceptron model
model = Sequential()

# RESHAPE is  working !!!!! >0<"
model.add(Reshape((look_back*nort,), input_shape=(look_back,nort)))
# Add layers to neruon network
model.add(Dense(190,input_dim=(look_back*nort), activation='relu'))
model.add(Dense(256, init='uniform', activation='relu'))
model.add(Dense(look_back, W_constraint=nonneg()))
model.compile(loss='mse', optimizer='adam')

model.fit(trainX, trainY, nb_epoch=168, batch_size=3, verbose=0)


# Estimate model performance
trainScore = model.evaluate(trainX, trainY, verbose=0)
print('Train Score: %.2f MSE (%.2f RMSE)' % (trainScore, math.sqrt(trainScore)))
testScore = model.evaluate(testX, testY, verbose=0)
print('Test Score: %.2f MSE (%.2f RMSE)' % (testScore, math.sqrt(testScore)))
    

```
